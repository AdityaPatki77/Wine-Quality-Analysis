```python
data.drop('total sulfur dioxide',axis = 1, inplace = True)
data.head()
```

```python
data.isna().sum()
```

```python
data.update(data.fillna(data.mean()))
```

```python
data.isna().sum()
```

```python
new_data = pd.get_dummies(data, drop_first=True)
new_data
```

```python
new_data['best quality'] = [ 1 if x >= 7 else 0 for x in data.quality]
new_data
```

```python
from sklearn.model_selection import train_test_split as tts

x = new_data.drop(['quality', 'best quality'], axis=1)
y = new_data['best quality']

x_train,x_test,y_train,y_test = tts(x,y,test_size = 0.2,random_state = 40)
```

```python
from sklearn.preprocessing import MinMaxScaler

norm = MinMaxScaler()
norm_fit = norm.fit(x_train)
new_x_train = norm_fit.transform(x_train)
new_x_test = norm_fit.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rnd = RandomForestClassifier()

fit_rnd = rnd.fit(new_x_train,y_train)

rnd_score = rnd.score(new_x_test,y_test)

rnd_score*100
```

```python
x_predict = list(rnd.predict(new_x_test))
predicted_data = {'predicted_values': x_predict, 'original_values': y_test}

pd.DataFrame(predicted_data).head(20)
```
